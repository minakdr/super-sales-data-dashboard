```python
# Check if there are negative values ( which isn't logic )
negative_values = df[['Sales', 'Quantity', 'Discount', 'Profit']] < 0
print(negative_values.any())
```

```python
#I want to rename some columns 
df = df.rename(columns={'Segment': 'Type_of_customer', 'Sales': 'Sales'})
```
